```python
import pandas as pd

citescorepct = pd.read_excel('citescore_2021.xlsx') # downloaded from scopus with ISSN search
```

```python
citescorepct.head
```

```python
percentile_lst = []
percentile_rank_lst = []
qualis_lst = []
percentile_area_lst = []
for str_pct in citescorepct['Highest percentile']:
    if isinstance(str_pct, str):
        str_out = str_pct.splitlines()
                                   
        pct = float(str_out[0][0:-1])
        rank = str_out[1]
        area = str_out[2]
        if (pct < 100) & (pct > 87.5):
            qualis = 'A1'
        elif (pct <= 87.5 ) & (pct > 75):
            qualis = 'A2'
        elif (pct <= 75 ) & (pct > 62.5):            
            qualis = 'A3'
        elif (pct <= 62.5 ) & (pct > 50):
            qualis = 'A4'
        elif (pct <= 50 ) & (pct > 37.5):
            qualis = 'B1'
        elif (pct <= 37.5 ) & (pct > 25):
            qualis = 'B2'
        elif (pct <= 25 ) & (pct > 12.5):
            qualis = 'B3'
        else:
            qualis = 'B4'
        
    else: # no info regarding citescore percentile available
        pct = 'NA'
        rank = 'NA'
        area = 'NA'
        qualis = 'NA'
        
    percentile_lst.append(pct)       
    percentile_rank_lst.append(rank)  
    qualis_lst.append(qualis)
    percentile_area_lst.append(area)

# percentile_lst    
# percentile_area_lst
```

```python
df_citescorepct = pd.DataFrame()

df_citescorepct['Journal'] = citescorepct['Source title']
df_citescorepct['CiteScore'] = citescorepct['CiteScore']
df_citescorepct['Highest Percentile'] = percentile_lst
df_citescorepct['Qualis CAPES'] = qualis_lst
df_citescorepct['Area'] = percentile_area_lst
df_citescorepct['Area rank'] = percentile_rank_lst
df_citescorepct['SNIP'] = citescorepct['SNIP']
df_citescorepct['SJR'] = citescorepct['SJR']

df_citescorepct.head
```

```python
# merge dataframes to add citescore data to all publications
df_pubs_citescore = pd.merge(left=df_pubs,right=df_citescorepct,left_on='publicationName',right_on='Journal')
# add year information in a collumn
df_pubs_citescore['Year'] = df_pubs_citescore['coverDate'].str.slice(stop=4).astype(int)
df_pubs_citescore['Year']
```

```python
# keep only journal papers
df_pubs_citescore = df_pubs_citescore[ df_pubs_citescore['subtype'] == 'ar' ]
df_pubs_citescore.head
```

```python
# get only columns of interest
df_print_pubs_citescore = df_pubs_citescore[['Year','Journal','CiteScore', 'Highest Percentile','Qualis CAPES', 'Area', 'Area rank', 'SNIP', 'SJR']] # nao cabe na folha
df_print_pubs_citescore = df_pubs_citescore[['Year','Journal','CiteScore', 'Highest Percentile','Qualis CAPES', 'SNIP', 'SJR']]
df_print_pubs_citescore = df_print_pubs_citescore.sort_values('Year',ascending=False,ignore_index=True)
df_print_pubs_citescore.head
```

```python
# remove 'Studies in Computational Intelligence' (book chapter)
df_print_pubs_citescore = df_print_pubs_citescore[0:-1]
df_print_pubs_citescore.index += 1 # comeca o indice em 1
print(df_print_pubs_citescore.to_latex())
```
